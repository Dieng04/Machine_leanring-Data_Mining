# Engineering missing values in numerical variables 

```python
# Missing data for categorical variables
print_missing_data(X_train, numerical)
```

* loser_age,loser_ht,loser_rank,loser_rank_points, winner_age, winner_ht, winner_rank and winner_rank_points contains < 50% NA: create additional variable with NA + random sample imputation

* l_svpt,loser_seed,w_svpt and winner_seed contains > 50% NA: impute NA by value far in the distribution

```python
def impute_na(X_train, df, variable):
    # make temporary df copy
    temp = df.copy()
        
    # extract random from train set to fill the na
    random_sample = X_train[variable].dropna().sample(temp[variable].isnull().sum(), random_state=0)
        
    # pandas needs to have the same index in order to merge datasets
    random_sample.index = temp[temp[variable].isnull()].index
    temp.loc[temp[variable].isnull(), variable] = random_sample
    return temp[variable].astype(float)
```

```python
# replace by random sampling
var1 =['loser_age','loser_ht','loser_rank','loser_rank_points', 'winner_age', 'winner_ht', 'winner_rank','winner_rank_points']
for df in [X_train, X_test]:
    for var in var1:
        df[var] = impute_na(X_train, df, var)
```

```python
var2 = ['l_svpt','loser_seed','w_svpt', 'winner_seed']
for df in [X_train, X_test]:
    for var in var2:
        extreme = X_train[var].mean() + X_train[var].std()*3
        df[var].fillna(extreme, inplace=True)
```

```python
# top coding and bottom coding for outiers according to interquantile proximity rule
upper_boundary ={}
lower_boundary={}
var_outlier = ['l_svpt','loser_age','loser_ht', 'loser_rank','loser_rank_points', 'w_svpt','winner_age','winner_ht','winner_rank','winner_rank_points']
for var in var_outlier:
    IQR = X_train[var].quantile(0.75)-X_train[var].quantile(0.25) 
    upper_b =  X_train[var].quantile(0.75) + (IQR * 3)
    lower_b= X_train[var].quantile(0.25)- (IQR * 3)
    upper_boundary[var]=upper_b
    lower_boundary[var]=lower_b
    
```

```python
upper_boundary
```

```python
lower_boundary
```

```python
lower_boundary['l_svpt']=0
lower_boundary['loser_rank']=0
lower_boundary['loser_rank_points']=0
lower_boundary['w_svpt']=0
lower_boundary['winner_rank']=0
lower_boundary['winner_rank_points']=0
```

```python
# update values: train and test
for var in var_outlier:
    #upper_fente
    X_train.loc[X_train[var]>upper_boundary[var], var] = upper_boundary[var]
    X_test.loc[X_test[var]>upper_boundary[var], var] = upper_boundary[var]
    #lower_fente
    X_train.loc[X_train[var]<lower_boundary[var], var] = lower_boundary[var]
    X_test.loc[X_test[var]<lower_boundary[var], var] = lower_boundary[var]
```

# Encode categorical variables


* Replace by risk probability

```python
X_test.isnull().mean()
```

```python
def encode_categorical_variables(var, target):
        # make label to risk dictionary
        ordered_labels = X_train.groupby([var])[target].mean().to_dict()
        
        # encode variables
        X_train[var] = X_train[var].map(ordered_labels)
        X_test[var] = X_test[var].map(ordered_labels)
```

```python
# enccode labels in categorical vars
X_train['won']=X_train["won"].astype('int')
X_test['won']=X_test["won"].astype('int')
for var in categorical:
    encode_categorical_variables(var, 'won')
```

* Evaluation criteria

Critères d’évaluation: Quand les effectifs des classes du jeu de données sont déséquilibrés, la métrique traditionnelle de mesure de performance de classification nommée Accuracy n’est pas adaptée. Dans notre les données sont parfaitement équilibrées comme nous pouvons le voir dans les figures ci-dessous.Donc nous utilisons Accuracy pour évaluer nos modèles.

* training classifiers

```python
# init time 
tim = time.time()
print('Learn. model\t\t score\t\t\ttime')
scores = []
y_test=y_test.astype('int')
y_train=y_train.astype('int')

for name, clf in zip(names, classifiers):
        print(name, end='')
        clf.fit(X_train, y_train.won)
        score = clf.score(X_test, y_test.won)
        print('\t\t', round(score, 3), '\t\t', round(time.time() - tim, 3))
        scores.append(score)
        tim = time.time()
```

* plot results

```python
# plot results
plt.rcdefaults()

y_pos = np.arange(len(names))

plt.bar(y_pos, scores, align='center', alpha=0.5, color='green')
plt.xticks(y_pos, names, rotation='vertical')
plt.ylabel('Accuracy')
plt.title('Model comparison for ATP prediction')

plt.show()
```
